### Oil Price Fluctuations

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create the data
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
prices_2021 = [60.82, 68.83, 73.47, 79.59]
prices_2022 = [100.30, 113.54, 100.71, 88.56]
prices_2023 = [81.17, 78.32, 86.66, 83.72]

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(quarters, prices_2021, marker='o', label='2021')
plt.plot(quarters, prices_2022, marker='s', label='2022')
plt.plot(quarters, prices_2023, marker='^', label='2023')

# Customize the plot
plt.title('Quarterly Oil Price Comparison (2021-2023)', pad=20)
plt.xlabel('Quarter')
plt.ylabel('Price')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

# Add value labels
for year_data in [prices_2021, prices_2022, prices_2023]:
    for i, price in enumerate(year_data):
        plt.annotate(f'{price:.2f}', 
                    (quarters[i], price),
                    textcoords="offset points",
                    xytext=(0,10),
                    ha='center')

plt.tight_layout()
plt.show()
```


Starting in 2021,the oil  pricesshowedd a consistent upward trajectory, climbing steadily from 60.82 to 79.59, marking a significant 30.9% increase over the year. However, 2022 emerged as a particularly turbulent yea withy dramatic price fluctuations. Prices surged to a peak of 113.54 in Q2 2022, representing the highest point in the three-year period, before beginning a steady decline.
